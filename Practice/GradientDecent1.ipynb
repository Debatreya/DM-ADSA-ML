```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y = make_regression(n_samples=10, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y, color='black')
```

```python
# OLS
import sklearn.linear_model as lm

reg = lm.LinearRegression()
reg.fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
def gradient_descent(X, y, m, b, lr, epochs):
    for i in range(epochs):
        loss_slope = -2 * np.sum(y - (m * X + b))
        step_size = lr * loss_slope
        b = b - step_size
    return b
```

```python
# Check after : epochs = 10
b = gradient_descent(X, y, m, b, lr, 10)
```
