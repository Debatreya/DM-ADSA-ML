```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

## Simple Linear Regression

```python
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=13, n_targets=1)
```

```python
plt.scatter(X, y)
```

```python
# Checking the original answer using sklearn
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X, y)
print(reg.coef_)
print(reg.intercept_)
```

```python
# Using my own implementation of the gradient descent
from GradD import GD_Regresser
```

```python
gd = GD_Regresser(0.01, 1000)
gd.fit(X, y)
```

```python
# Plotting the line
plt.scatter(X, y)
plt.plot(X, gd.predict(X), color='red', label='My GD')
plt.plot(X, reg.predict(X), color='green', label='Sklearn')
plt.show()
```

```python
# Checking the answer
print(gd.m)
print(gd.b)
print(reg.coef_)
print(reg.intercept_)

# Checking the loss
print(reg.score(X, y))
```
