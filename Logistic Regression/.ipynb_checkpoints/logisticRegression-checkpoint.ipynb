# Logistic Regression

### Importing the required libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

### Loading the data frames

```python
df = pd.read_csv('./bank/bank.csv', sep=';', header=0)
df
```

### One-Hot Encoding (of categorical data)
One-Hot Encoding is a method of representing characters or words by a vector where only one element is set to one and all others are zero, based on their position in the vocabulary. This results in a sparse, semantically independent vector with a high dimension.

```python
data = pd.get_dummies(df, drop_first=True)
data
```

### Separate Dependent Varables (Y) and Independent Variables (X)

```python
X = data.drop('y_yes', axis=1)  # 'y_yes' is the encoded target column after one-hot encoding
Y = data['y_yes']
print(X)
print(Y)
```

### Converting the X and Y to numpy arrays

```python
# Convert to numpy arrays
x = X.values
y = Y.values

x = np.array(x, dtype=int)

print(x, y)
```

### Normalize / Standardise features (calculatind z-score of normal distribution)
$$z= (Xâˆ’Î¼) / Ïƒ$$


```python
mean = x.mean(axis=0)
std = x.std(axis=0)

x = (x - mean) / std

x
```

### Initialize Parameters

```python
m, n = x.shape
beta = np.zeros(n)
bias = 0
learning_rate = 0.01
iterations = 5000
```

### Sigmoid Function

```python
def sigmoid(z):
    # Ensure z is a NumPy array and convert to float
    z = np.array(z, dtype=float)
    
    # Clip values to avoid overflow/underflow
    z = np.clip(z, -500, 500)
    
    return 1 / (1 + np.exp(-z))
```

### Logistc Regression Model

```python
def logistic_regression(X, y, beta, bias, learning_rate, iterations):
    m = X.shape[0]
    # Cost iteration points for graph plotting
    costArray = []
    learning_rate = float(learning_rate)
    
    for i in range(iterations):
        # Compute the linear combination of inputs and weights
        z = np.dot(X, beta) + bias
        # Apply the sigmoid function
        y_hat = sigmoid(z)
        
        # Compute the cost (log-loss)
        cost = -1/m * np.sum(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
        costArray.append(cost)
        
        # Compute gradients
        dbeta = 1/m * np.dot(X.T, (y_hat - y))
        dbias = 1/m * np.sum(y_hat - y)

        beta = np.array(beta, dtype=np.float64)
        dbeta = np.array(dbeta, dtype=np.float64)
        bias = np.array(bias, dtype=np.float64)
        dbias = np.array(dbias, dtype=np.float64)
        
        # Update parameters using gradient descent
        beta -= learning_rate * dbeta
        bias -= learning_rate * dbias
        
        # Optional: Print the cost every 100 iterations
        # if i % 100 == 0:
        #     print(f'Cost after iteration {i}: {cost}')
    
    return beta, bias, costArray
```

### Train Model

```python
beta, bias, cost = logistic_regression(X, y, beta, bias, learning_rate, iterations)
```

### Calculated Bias and Beta

```python
print("Bias: ", bias)
print("Beta: ", beta)
print("Logistic regression (y) = ", sigmoid(np.dot(x, beta) + bias))
```

### Make Predictions

```python
def predict(X, beta, bias, threshold=0.5):
    y_hat = sigmoid(np.dot(X, beta) + bias)
    return [1 if i > threshold else 0 for i in y_hat]
```

### Predict on training set

```python
predictions = predict(x, beta, bias)
```

### Check accuracy (Evaluate Model)

```python
def accuracy(y, y_pred):
    return np.mean(y == y_pred)
```

### Calculate Accuracy

```python
acc = accuracy(y, predictions)
print(f'Accuracy: {acc * 100:.2f}%')
```

### Visualize the cost over iterations

```python
%matplotlib inline
plt.plot(cost, color='blue')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost reduction over iterations')
plt.show()
```
