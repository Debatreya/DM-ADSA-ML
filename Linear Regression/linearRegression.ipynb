# Linear Regression

## Importing libraries
- [x] pandas
- [x] numpy
- [x] mathplotlib.pyplot
- [x] linear molde from `sklearn`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

## Loading Dataset using pandas

```python
df = pd.read_csv('./LR.csv')
df
```

## Initial plot of the points

```python
%matplotlib inline
plt.xlabel('Income')
plt.ylabel('Food Expenditure')
plt.scatter(df.Income, df["Food Expenditure"], color='red', marker='+')
```

## Linear Regression using sicket learn (linear model)

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Income']], df["Food Expenditure"])
```

## Slope of line (m)
## Intercept (c)

```python
m = reg.coef_[0]
c = reg.intercept_
# Print linear equation
print('y = m*x + c')
print('y =', m, '* x +', c)
```

## Plot the points and Linear Regression

```python
%matplotlib inline
plt.xlabel('Income')
plt.ylabel('Food Expenditure')
plt.scatter(df.Income, df["Food Expenditure"], color='red', marker='+')
plt.plot(df.Income, reg.predict(df[['Income']]), color='blue', label='Best fitted line')
```
